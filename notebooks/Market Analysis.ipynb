```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("campaign_data.csv")
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert numeric columns to proper types
numeric_cols = ['impressions', 'clicks', 'spent', 'total_conversion', 'approved_conversion']
for col in numeric_cols:
    df[col] = df[col].astype(int) if col != 'spent' else df[col].astype(float)

# Create a revenue proxy (assuming $100 per approved conversion)
df['revenue_estimated'] = df['approved_conversion'] * 100

# Now revenue_estimated is numeric
df['revenue_estimated'] = df['revenue_estimated'].astype(float)
```

AGGREGATE BY XYZ CAMPAIGN

```python
campaign_metrics = df.groupby('xyz_campaign_id').agg({
    'spent':'sum',
    'clicks':'sum',
    'impressions':'sum',
    'approved_conversion':'sum',
    'revenue_estimated':'sum'
}).reset_index()
```

```python
# Calculate KPIs
campaign_metrics['ctr'] = campaign_metrics['clicks'] / campaign_metrics['impressions'] * 100
campaign_metrics['conversion_rate'] = campaign_metrics['approved_conversion'] / campaign_metrics['clicks'] * 100
campaign_metrics['roi'] = (campaign_metrics['revenue_estimated'] - campaign_metrics['spent']) / campaign_metrics['spent'] * 100
campaign_metrics['cost_per_conversion'] = campaign_metrics['spent'] / campaign_metrics['approved_conversion']

campaign_metrics.head()
```

ROI per Campaign

```python
plt.figure(figsize=(8,5))
sns.barplot(x='xyz_campaign_id', y='roi', data=campaign_metrics)
plt.title("ROI per XYZ Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("ROI (%)")
plt.xticks(rotation=45)
plt.show()
```

Conversions vs Spend (Campaign-Level)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='spent', y='approved_conversion', data=campaign_metrics, s=100)
plt.title("Conversions vs Spend by Campaign")
plt.xlabel("Spend (USD)")
plt.ylabel("Approved Conversions")
plt.show()
```

CTR Distribution

```python
plt.figure(figsize=(8,5))
sns.barplot(x='xyz_campaign_id', y='ctr', data=campaign_metrics)
plt.title("CTR by Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("CTR (%)")
plt.xticks(rotation=45)
plt.show()
```
